Q1 What is a Common Table Expression (CTE), and how does it improve SQL query readability?

A) Common Table Expression (CTE) is a temporary named result set in SQL that is defined using the WITH clause and can be used within a SELECT, INSERT, UPDATE, or DELETE statement.

*Example:*

        WITH cte_name AS (
    SELECT column1, column2
    FROM table_name
    WHERE condition
    )
    SELECT *
    FROM cte_name;

**improves readability**

 **1. Breaks complex queries into smaller parts** – Makes long queries easier to understand.

**2. Reusable within the same query** – You can reference the CTE multiple times.

**3. Improves maintenance** – Easier to modify logic compared to nested subqueries.

**4. Better structure** – Makes SQL code cleaner and organized.

*Example:*

        WITH HighSalaryEmployees AS (
    SELECT name, salary
    FROM Employees
    WHERE salary > 50000
    )
    SELECT name
    FROM HighSalaryEmployees;

Q2 Why are some views updatable while others are read-only? Explain with an example.

A) Some views are updatable because they are based on a single table and do not contain complex SQL features. In such cases, changes made through the view can be directly reflected in the original table.

Some views are read-only because they use joins, aggregate functions (SUM, COUNT), GROUP BY, DISTINCT, or calculated columns, making it unclear how updates should be applied to base tables.

**View Examples:**

      CREATE VIEW Employee_View AS
    SELECT EmpID, Name, Salary
    FROM Employees;

  
      UPDATE Employee_View
    SET Salary = 60000
    WHERE EmpID = 101;


**Read-Only Examples**

    CREATE VIEW Dept_Salary AS
    SELECT DeptID, AVG(Salary) AS AvgSalary
    FROM Employees
    GROUP BY DeptID;

This view is read-only because AVG(Salary) is derived data, so direct updates are not possible.

Q3  What advantages do stored procedures offer compared to writing raw SQL queries repeatedly?

A) Stored procedures offer several advantages over writing raw SQL queries repeatedly:

1. Reusability – Write once and use many times.

2. Better Performance – Stored procedures are precompiled, so execution is faster.

3. Reduced Code Duplication – Avoid repeating the same SQL queries in applications.

4. Security – Users can be given permission to execute procedures without direct table access.

5. Easy Maintenance – Changes can be made in one place instead of updating multiple queries.

6. Consistency – Ensures the same business logic is used everywhere.

**Examples:**

      CREATE PROCEDURE GetEmployeeDetails
    AS
    SELECT * FROM Employees;

  **Instead of writing:**

    SELECT * FROM Employees;

  **repeatedly, you can simply call:**

    EXEC GetEmployeeDetails;

Q4  What is the purpose of triggers in a database? Mention one use case where a trigger is essential.

A) A trigger is a special stored program in a database that automatically executes when certain events occur on a table, such as INSERT, UPDATE, or DELETE.

**Purpose of Triggers:**

**1.***Automatically enforce business rules*

**2.** *Maintain data integrity*

**3.** *Audit changes by storing logs*

**4.** *Synchronize related tables*

**5.** *Prevent invalid operations*

**Essential Use Case:**

***Audit Logging*** – When an employee record is deleted, a trigger can automatically store the deleted record in a backup/audit table.

    CREATE TRIGGER trg_DeleteEmployee
    AFTER DELETE ON Employees
    FOR EACH ROW
    INSERT INTO Employee_Audit(EmpID, Name)
    VALUES (OLD.EmpID, OLD.Name);

Q5 Explain the need for data modelling and normalization when designing a database.

A) **Data modelling** and **normalization** are important steps in database design because they help organize data efficiently and avoid future problems.

**Need for Data Modelling:**

Data modelling is the process of defining tables, relationships, and attributes before creating the database.

**Benefits:**

- Gives a clear structure of the database

- Identifies relationships between entities

- Reduces design errors early

- Makes development and maintenance easier

**Need for Normalization:**

Normalization is the process of organizing data into well-structured tables to reduce redundancy and dependency.

**Benefits:**

- Eliminates duplicate data

- Improves data consistency

- Prevents update, insert, and delete anomalies

- Saves storage space

- Makes database easier to manage

**Example:**

Instead of storing customer and order details repeatedly in one table, normalization separates them into:

- Customers table

- Orders table

linked by CustomerID.

This keeps data clean, accurate, and efficient.
