Q9.  What is the Difference Between a Primary Key and a Foreign Key?

A)  **- Primary Key**

  - Definition: A column (or set of columns) that uniquely identifies each record in a table.
- Must be unique for every row.
- Cannot contain NULL values.

   Example:

     CREATE TABLE Employees (
    emp_id INT PRIMARY KEY,
    name VARCHAR(100),
    department VARCHAR(50)
);

    - Here, emp_id is the primary key.


  **-Forgien Key**

    A column (or set of columns) in one table that refers to the primary key in another table.

    Maintains referential integrity between tables,ensuring that relationships are valid

    Example:

        CREATE TABLE Sales (
    sale_id INT PRIMARY KEY,
    emp_id INT,
    amount DECIMAL(10,2),
    FOREIGN KEY (emp_id) REFERENCES Employees(emp_id)
);

  Here, emp_id in saless is a forgien key referencing primary key emp_id in Employees.

Q10.  What Are Constraints in SQL and Why Are They Used?

A)  Constraints in SQL are rules used to limit the type of data that can be stored in a table, ensuring the accuracy and reliability of the information within a database. They control what type of data can be inserted, updated, or deleted, ensuring that the database always follows business logic and remains consistent.

 Why Constraints Are Used:

 - Ensure data validity (e.g., no negative salaries, no duplicate IDs).
- Prevent errors by rejecting invalid operations.
- Maintain relationships between tables (e.g., orders must link to valid customers).
- Improve reliability by enforcing business rules automatically.
- Reduce redundancy and inconsistencies in stored data.

PRIMARY KEY:

      - Ensures each row is uniquely identifiable; no duplicates, no NULLs
      - emp_id INT PIMARY KEY

FOREIGN KEY:

      - Enforces referential integrity between tables
      - FOREIGN KEY (dept_id) REFERENCE

NOT NULL:

      - Prevents NULL values in a column
      - name VARCHAR(50) NOT NULL


UNIQUE:

      - Ensures all values in a column are distinct
      - email VARCHAR(100) UNIQUE

 CHECK :  

        - Validates values against a condition
        - salary DECIMAL CHECK (salary > 0)

        **ETC**

Example in Practice:

      CREATE TABLE Employees (
    emp_id INT PRIMARY KEY,
    name VARCHAR(100) NOT NULL,
    email VARCHAR(100) UNIQUE,
    salary DECIMAL(10,2) CHECK (salary > 0),
    hire_date DATE DEFAULT CURRENT_DATE
);
