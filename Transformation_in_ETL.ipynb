Q1  Define Data Transformation in ETL and explain why it is important

A) Data Transformation in ETL is the process of converting raw, inconsistent data into a clean, structured, and usable format. It is important because it ensures accuracy, consistency, and relevance of data, enabling reliable analytics, reporting, and business decision-making.

- Data Transformation is the second stage of the ETL (Extract, Transform, Load) pipeline.

-  It involves cleaning, restructuring, and enriching extracted data so it can be loaded into a target system (like a data warehouse or data lake).

- Common operations include:

  - Data cleansing: Removing duplicates, fixing errors, handling missing values.

  - Data standardization: Converting formats (e.g., date formats, currency units).

  - Data integration: Merging data from multiple sources into a unified schema.

  - Data enrichment: Adding derived values or external information to make data more useful.

#Importance of Data Transforming

1. Ensures Data Quality

- Raw data often contains errors, inconsistencies, or missing values.

- Transformation improves accuracy, completeness, and reliability.

2. Enables Data Integration


- Different systems store data in different formats. Transformation harmonizes these formats, creating a single, unified view across sources.

3. Supports Business Intelligence & Analytics


- Clean, structured data is essential for dashboards, reports, and predictive models. Without transformation, insights would be misleading.

4. Improves Decision-Making

- Reliable data allows organizations to make informed, evidence-based decisions. Transformation ensures that executives and analysts trust the data.

5. Compliance & Standardization

- Many industries require data to follow strict standards (e.g., financial reporting, healthcare records). Transformation enforces these rules.

**Key Takeaway**

Data transformation is the heart of ETL—without it, extracted data remains messy and unusable. By cleaning, standardizing, and enriching data, organizations can unlock accurate insights, better decisions, and competitive advantage.

Q2 List any four common activities involved in Data Cleaning.

A) Here are four common activities involved in Data Cleaning that are essential in ETL pipelines:

1.  Removing Duplicates

    - Identifying and eliminating repeated records to avoid double-counting or skewed analysis.
    - Example: Two identical customer entries in a CRM system.

2. Handling Missing Values

    - Filling in gaps with default values, averages, or removing incomplete records depending on the context.
    - Example: Replacing missing age values with the median age of the dataset.

3. Correcting Inconsistencies

    - Standardizing formats, units, and naming conventions to ensure uniformity.

    - Example: Converting all date formats to YYYY-MM-DD or ensuring "USA" and "United States" are treated as the same.

4. Validating Data Accuracy

    - Checking that values fall within expected ranges or match business rules.

    - Example: Ensuring product prices are non-negative or email addresses follow proper syntax.

Q4  A dataset has missing values in the “Age” column. Suggest two techniques to handle this and explain when they should be used.

A)
#1. Imputation with Statistical Measures (Mean/Median/Mode)
- How it works: Replace missing ages with the mean, median, or mode of the existing values.

- When to use:

- Mean: Suitable when the data is normally distributed and outliers are minimal.

- Median: Preferred when the data is skewed or contains outliers (e.g., a few very high ages).

- Mode: Useful for categorical-like age groups (e.g., age bands).

- Example: If most ages are between 20–30 but a few entries are missing, filling them with the median age (say 25) maintains balance.

#2. Predictive Imputation (Using Regression or Machine Learning Models)
- How it works: Use other features (e.g., income, education, occupation) to predict missing ages through regression, decision trees, or more advanced models.

- When to use:

- When the dataset is large and other variables strongly correlate with age.

- Useful in business or research contexts where accuracy of age is critical.

- Example: If younger employees tend to have lower salaries and fewer years of experience, a regression model can estimate missing ages based on those attributes.

Q6 What is One-Hot Encoding? Give an example with the categories: “Red, Blue, Green”.

A) One-Hot Encoding is a technique used to convert categorical variables into a numerical format that machine learning models can understand. Instead of assigning arbitrary numbers (which could imply an order), one-hot encoding creates binary columns for each category, marking presence with 1 and absence with 0.

Example with Categories: “Red, Blue, Green”
Suppose we have a dataset with the color column:

Original Data:  
["Red", "Blue", "Green", "Red"]

After One-Hot Encoding:
| Color | Red | Blue | Green |
| --- | --- | --- | --- |
| Red | 1 | 0 | 0 |
| Blue | 0 | 1 | 0 |
| Green | 0 | 0 | 1 |
| Red | 1 | 0 | 0 |

**Why It’s Important:**

- Prevents models from misinterpreting categories as having numerical order (e.g., “Red=1, Blue=2, Green=3” would wrongly suggest Green > Blue > Red).

-  Makes categorical data usable in algorithms like regression, decision trees, and neural networks.

- Ensures each category is treated independently.

Q7  Explain the difference between Data Integration and Data Mapping in ETL.

A)
#Data Integration
- Definition: The process of combining data from multiple sources (databases, APIs, flat files, etc.) into a unified view.

- Goal: To create a single, consistent dataset that can be used for analysis or reporting.

- Example: Merging customer records from a CRM system, sales data from an ERP, and feedback from a web app into one consolidated warehouse table.

- Key Point: Integration is about bringing data together across systems.

#Data Mapping
- Definition: The process of defining relationships between source fields and target fields during ETL.

- Goal: To ensure that data from different sources is correctly aligned with the schema of the destination system.

- Example: Mapping Cust_ID in the source database to CustomerID in the warehouse, or mapping DOB (date of birth) to Birth_Date.

- Key Point: Mapping is about establishing rules and correspondences so integration can happen correctly.
