4) Explain the difference between Sorting and Filtering with an example from this dataset.
A)  
    **SORTING**
    Defifnition:  Sorting arranges data in a particular order (ascending or decending) based on a column.
    Example:
          The sales column sort in descending order, the highest sales (497,483,467,etc) are appear at the top, and the lowest (53,76, 205,etc) at the bottom.
    This helps you quickly identify the best and worst sales values across all regions.

    **FILLTERING**
    Definition:
           Filtering hides rows that dont meet specific criteria, showing only the data you want to analyze.
    Example:
            The region column filter to show only South, you will see only the rows where Region=south (e.g., Oranges 348, Mangoes 247, Aples 417, etc).
    This helps you focus on sales performance in one region without deleting other data.

    .Data Visualization:
                  Dashboards turn complex data into charts, graphs, and KPIs, making trends and issues easy to spot.

    Real-Time Monitoring:
                They allow managers to track performance as it happens, rather than relying only on past reports.
