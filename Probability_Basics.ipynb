Q1) A die is rolled. What is the probability of getting:

    (a) An even number

    (b) A number greater than 4

A)  A fair die has 6 faces: {1,2,3,4,5,6}.
Each outcome is equally like,
So, the probability of any single face is 6.

(a) probability of getting an even number
- Even numbers on a die: {2,4,6}
- Number of favorable outcomes = 3
- Total outcomes = 6
    P(even number) = 3/6 = 1/2

(b) Probability of getting a number greater than 4

- Numbers greater than 4: {5, 6}
- Number of favorable outcomes = 2
- Total outcomes = 6

    P(number > 4) = 2/6 = 1/3

Answer is:

(a) 1/2

(b) 1/3

Q4. The population of a school is divided into 60% boys and 40% girls. If you want equal representation
of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling?
Why?

A)
 - Simple Random Sampling (SRS): Every student has an equal chance of being selected. Since the school population is 60% boys and 40% girls, an SRS will likely reflect this imbalance in the sample too. That means boys will be overrepresented compared to girls.

 -  Stratified Sampling: The population is divided into strata (here, boys and girls). Then you select equal numbers (or proportional numbers, depending on the goal) from each stratum. If you want equal representation of both genders, you deliberately sample the same number of boys and girls.

- Stratified Sampling ensures fairness and balance in representation, while Simple Random Sampling would mirror the population’s natural proportions.

Q8.  From a deck of 52 cards, two cards are drawn without replacement. What is the probability that
both are Aces?

A)

1. Probability of drawing the first Ace:

There are 4 Aces in a deck of 52 cards.

P(1st card is Ace) = 4/52 = 1/13

2. Probability of drawing the second Ace (given the first was an Ace and without replacement):

After drawing one Ace, there are 3 Aces left and a total of 51 cards remaining.

P(2nd card is Ace | 1st card was Ace) = 3/51 = 1/17

3. Probability that both cards drawn are Aces:

P(both are Aces) = P(1st is Ace) * P(2nd is Ace | 1st was Ace)

P(both are Aces) = (4/52) * (3/51)

P(both are Aces) = (1/13) * (1/17)

P(both are Aces) = 1/221

Final Answer: The probability that both cards drawn are Aces is 1/221.

Q9. A factory produces bulbs with 2% defective rate. If 5 bulbs are chosen at random, what is the
probability that all are non-defective?

A)

1. Probability of one bulb being non-defective:

- Defective rate = 2% = 0.02

- Non-defective rate = 1 - 0.02 = 0.98

2. Probability that all 5 bulbs are non-defective:

- Since each bulb's defectiveness is independent, we multiply the probability of a single bulb being non-defective by itself 5 times.

- P(all 5 non-defective) = (0.98) ^ 5
- P(all 5 non-defective) = 0.9039207968

Final Answer: The probability that all 5 chosen bulbs are non-defective is approximately 0.9039 or 90.39%.

Q10. Differentiate between discrete and continuous random variables with examples.

A) Discrete vs. Continuous Random Variables

- Definition: A discrete random variable is a variable whose value can only take a countable number of distinct values. These values are often integers and are typically the result of counting.

- Characteristics:
    - Can be counted.
    - Usually takes on integer values.
    - There are gaps between possible values.

- Examples:
  - The number of heads when flipping a coin 3 times (possible values: 0, 1, 2, 3).

  - The number of cars passing a specific point on a road in an hour.

  - The number of defective items in a batch of 100 products.
  
  - The number of students in a classroom.

- Continuous Random Variable:

 - Definition: A continuous random variable is a variable whose value can take any value within a given range or interval. These values are typically the result of measuring.
  
- Characteristics:
  - Cannot be counted; instead, they are measured.
  - Can take on any value within a continuous interval (including decimals and fractions).
  - There are no gaps between possible values.

- Examples:

  - The height of a person (e.g., 170 cm, 170.5 cm, 170.55 cm).

  - The weight of an apple.

  - The time it takes to complete a race.

  - The temperature in a room.

  - The amount of rainfall in a city during a month.
