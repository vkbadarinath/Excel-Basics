1) Explain the difference bitween Relative, Absolute and Mixed Cell Referencing.

    *   Relative Reference:
          A relative reference changes automatically when you copy the formula to another cell.
      
    *   Absolute Reference:
          An absolute reference does not change when copied.
          Dollar ($) sign is used before column and row.
          Both row and column are locked.
  
    *   Mixed Reference:
          Only one part (row or column) is fixed.
        There are two types:
          $A1 Column fixed, row changes
          A1 Row fixed, column changes Example: A1*B$1
          Either row or column is locked.
          Used when: Creating multiplication tables or structured data tables.
