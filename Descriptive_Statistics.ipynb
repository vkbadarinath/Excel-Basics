Q1: Understanding Central Tendency

A bakery tracks the daily sales of muffins (in dozens) over a week: [10, 12, 11, 15, 14, 13, 12].

What is the most representative value of their weekly sales, and why?

A) The bakery’s weekly muffin sales (in dozens) are:
[10, 12, 11, 15, 14, 13, 12]

- Mean (Average):
{10+12+11+15+14+13+12} = 87/7 (12.43)

- Median (Middle Value):

Ordered data: [10, 11, 12, 12, 13, 14, 15]
Middle value = 12

- Mode (Most Frequent Value):

The value 12 occurs twice, more than any other.

The most representative value is 12, both the median and mode point to 12, and the mean is very close to it. This consistency makes 12 the best summary of the bakery’s typical daily sales.

Q4: Median in Real Life.

A car dealer notes the prices of used cars: [$8,000, $9,500, $10,200, $11,000, $50,000]

Why is the median a better measure than the mean in this case? Calculate the median.

A) Car prices:
[$8,000, $9,500, $10,200, $11,000, $50,000]

- Arrange in order (already sorted):
[$8,000, $9,500, $10,200, $11,000, $50,000]
- Middle Value (3rd out of 5) = $10,200

Median = $10,200

- Mean (Average):
{8,000+9,500+10,200+11,000+50,000} = 88,700 / 5 = (17,740)


- So, mean = $17,740

- Notice:  
        how the $50,000 outlier pulls the mean way up, making it look like the “average” car costs nearly $18,000.

- The median ($10,200) reflects the “typical” car price without being distorted by the one unusually high value.
- The mean ($17,740) exaggerates the average because of the outlier, giving a misleading impression of affordability.

The median is $10,200, and it’s a better measure than the mean because it accurately represents the central tendency of car prices without being skewed by the $50,000 outlier.

Q6: Range in Action

A farmer records the weekly weight of harvested apples (kg): [100, 105, 98, 110, 120].

Find the range. How can this help the farmer in planning his packaging?

A) Apple harvest weights (kg):
[100, 105, 98, 110, 120]

Calculate the Range

Range = Maximum - Minimum =120-98=22


 The range = 22 kg

- A 22 kg difference means the farmer’s harvest varies quite a bit from week to week.
- This variation is important for packaging and logistics:
- If the farmer prepares packaging only for the smallest harvest (98 kg), they’ll run short when the harvest is larger (up to 120 kg).
- If they prepare for the largest harvest every time, they may waste packaging materials when the harvest is smaller.


Q7: Variance for Decision-Making

Two delivery companies track delivery delays (in minutes).

Company A: variance = 6

 Which company is more consistent, and why?

 A)  
 - Company A: variance = 6
- Company B: variance = 15


- A smaller variance means the values are closer to the average → more consistent performance.
- A larger variance means the values fluctuate more → less predictable.

- Company A’s variance (6) is lower than Company B’s (15).
- This means Company A’s delivery delays are more tightly clustered around the average delay, showing greater consistency.
- Company B’s higher variance indicates delays vary more widely, making them less reliable.
