Q1. What is a null hypothesis (H₀) and why is it important in hypothesis testing?

A) A null hypotheesis (Ho) is a statement that there is no effect, no relationship, or no difference between two or more variablesbeing tested. It represents the default assumpstion or the status quo.

It is crucial in hypothesis testing because:

  **Key Points:**

  **. Definition:**  Ho is the statement that there is no significant effect or association.
  
  *Example:* "There is no ifference in test scores between students who study with music and those who study in silence."

  **.Role in Testing:** It serves as the baseline aginst which the alternative hypothesis (H1 or Ha) is tested.
    
  **.Decision Making:** Statistical tests (like t-tests, chi-square tests, ANOVA) calculate probabilities (p-values) to determine whether the data provide enough evidence to reject Ho.

  **.Objectivity:** By starting with the assumption of "no effect", researchers avoid bias and ensure that only strong evidence leads to claims ofa real efffect.

  **Why It's Important**

  **.provides a benchmark:**  Without Ho, we wouldn't have a clear reference point to evaluate results.

  - Controls false claims: It reduces the risk of concluding that an effect exists when it doesn’t (Type I error).
- Scientific rigor: It enforces a structured, evidence-based approach to research, ensuring findings are credible and reproducible.

Q2.  What does the significance level (α) represent in hypothesis testing?

A) The significance level (α) is a threshold used in hypothesis testing to decide whether to reject the null hypothesis (H₀). It represents the probability of making a Type I error—rejecting H₀ when it is actually true.

** Key Points**

- Definition: α is the maximum risk you are willing to accept of wrongly concluding that an effect exists.
- Common Values: Researchers often set α at 0.05 (5%), 0.01 (1%), or 0.10 (10%).
- Interpretation:
- If the p-value ≤ α → reject H₀ (evidence suggests a real effect).
- If the p-value > α → fail to reject H₀ (not enough evidence to claim an effect).
- Balance of Errors: Lower α reduces the chance of false positives (Type I error) but increases the chance of false negatives (Type II error).

Q3.  Differentiate between Type I and Type II errors.

A) **Type I Error**

- Definition: Rejecting the null hypothesis (H₀) when it is actually true.
- Nickname: "False positive."
- Example: A medical test says a patient has a disease when they actually don’t.
- Probability: Equal to the significance level (α).
🔵 Type II Error
- Definition: Failing to reject the null hypothesis when the alternative hypothesis (H₁) is true.
- Nickname: "False negative."
- Example: A medical test says a patient is healthy when they actually have the disease.
- Probability: Denoted by β. The power of a test is 1-\beta , which measures the ability to detect a real effect.

**Comparison Table**

- Error Type I

  - Reject Ho when Ho is true

  - False positive
  - Probability (a)
  - Declaring a drug effective when it isn't

- Error Type II

  - Fail to reject Ho when H1 is true
  - False positive
  - Probability(B)
  - Missing that a drug truly works

**Why It Matters**

- *Type I error* - risk spreading false claims (e.g., approving ineffective treatments).
- *Type II errors* -  risk overlooking real discoveries (e.g., missing a life-saving drug).

- Researchers must balance α and β depending on the context—medical trials often prioritize minimizing Type I errors, while exploratory studies may tolerate them to reduce Type II errors.

Q4. Explain the difference between a one-tailed and two-tailed test. Give an example of each.

A) **One-Tailed Test**
- Definition: Tests for an effect in one specific direction (greater than OR less than).
- Use Case: When you have a directional hypothesis.
- Example:
- A company claims a new drug increases recovery speed compared to the standard drug.
- Null hypothesis (H₀): New drug recovery time ≤ Standard drug recovery time.
- Alternative hypothesis (H₁): New drug recovery time > Standard drug recovery time.
- Here, you only care if the new drug is better, not worse.
🔹 **Two-Tailed Test**
- Definition: Tests for an effect in both directions (different, whether greater OR less).
- Use Case: When you don’t know the direction of the effect, only that a difference may exist.
- Example:
- A researcher wants to know if a new teaching method affects test scores compared to traditional teaching.
- Null hypothesis (H₀): No difference in scores.
- Alternative hypothesis (H₁): Scores are different (could be higher OR lower).
- Here, you’re open to finding improvement or decline.

Q5. A company claims that the average time to resolve a customer complaint is 10 minutes.

A random sample of 9 complaints gives an average time of 12 minutes and a standard deviation of 3 minutes.

At α = 0.05, test the claim

A) **Problem Setup**
- Claimed mean (μ₀): 10 minutes
- Sample mean $\bar{x}$:12 minutes
- Sample standard deviation (s): 3 minutes
- Sample size (n): 9
- Significance level (α): 0.05

We’ll use a t-test since the sample size is small (n < 30) and population standard deviation is unknown.

🔹  State Hypotheses
- Null hypothesis (H₀): μ = 10 (average resolution time is 10 minutes)
- Alternative hypothesis (H₁): μ ≠ 10 (average resolution time is different from 10 minutes)

-  This is a two-tailed test.

🔹Test Statistic
Formula:

  $t=\frac{\bar {x}-\mu _0}{s/\sqrt{n}}$

Substitute values:

$t=\frac{12-10}{3/\sqrt{9}}=\frac{2}{3/3}=\frac{2}{1}=2$

So, t=2.0.

🔹 Step 3: Degrees of Freedom

  df=n-1=9-1=8

🔹 Step 4: Critical Value
For a two-tailed test at α = 0.05 with df = 8:

  $t_{critical}\approx \pm 2.306$

🔹 Step 5: Decision Rule
- If  $|t|>t_{critical}, reject H₀.$
- Here, $|t|=2.0$  which is less than 2.306.

Q6.  When should you use a Z-test instead of a t-test?

A)
- You should use a Z-test when:

The population standard deviation (σ) is known.

The sample size is large (typically n ≥ 30), so the Central Limit Theorem ensures the sampling distribution of the mean is approximately normal.

The data are normally distributed (or close enough for large samples).

- In contrast, use a t-test when:

The population standard deviation is unknown (you only have the sample standard deviation, s).

The sample size is small (n < 30).

The data are approximately normal (important for small samples).

Q7. The productivity of 6 employees was measured before and after a training program

A)  This is a paired sample t-test problem since we’re comparing productivity before and after training

- Null hypothesis (H₀): Training has no effect$ → μd = 0$
- Alternative hypothesis (H₁): Training improves productivity → μd > 0

 - Mean difference:

$\bar {d}=\frac{5+5+1+3+1+3}{6}=\frac{18}{6}=3$

- Variance:

$s_d^2=\frac{\sum (d_i-\bar {d})^2}{n-1}$


**Deviations:** (5-3)=2, (5-3)=2, (1-3)=-2, (3-3)=0, (1-3)=-2, (3-3)=0
Squares: 4, 4, 4, 0, 4, 0 → Sum = 16

$s_d^2=\frac{16}{5}=3.2\quad \Rightarrow \quad s_d=\sqrt{3.2}\approx 1.789$


 **Test Statistic**

$t=\frac{\bar {d}-0}{s_d/\sqrt{n}}=\frac{3}{1.789/\sqrt{6}}$

$t=\frac{3}{1.789/2.449}=\frac{3}{0.730}\approx 4.11$



**Critical Value**

- Degrees of freedom: df=n-1=5
- One-tailed test at α = 0.05$ → t_{critical}\approx 2.015$

 **Decision**
- Calculated t=4.11
- Critical t=2.015
- Since $t>t_{critical}, reject H₀.$

Q8. A company wants to test if product preference is independent of gender.

A) Step 1: State Hypotheses
- Null hypothesis (H₀): Product preference is independent of gender.
- Alternative hypothesis (H₁): Product preference depends on gender.

Step 2: Observed Frequencies (O)

From the table:

  Gender:- Male,

    Product A:- 30,

    Product-B:- 20,

    Total:- 50

  Gender:- Female,

    Product A:- 10,

    Product B:- 40,

    Total:- 50,

  Total:-

    Product A:- 40,

    Product B:- 60,

    Total:- 100,

  
Step 3: Expected Frequencies (E)

Formula:

$E_{ij}=\frac{(\mathrm{Row\  Total})(\mathrm{Column\  Total})}{\mathrm{Grand\  Total}}$

- Male, Product A: $\frac{50\cdot 40}{100}=20$

- Male, Product B: $ \frac{50\cdot 60}{100}=30$

- Female, Product A: $\frac{50\cdot 40}{100}=20$

- Female, Product B: $\frac{50\cdot 60}{100}=30$

  So expected table:

  Gender:- Male,

      Product A:- 20,

      Product B:- 30,

  Gender:- Female,

      Product A:- 20,

      Product B:- 30,

  
Step 4: Chi-Square Test Statistic

Formula:

$\chi ^2=\sum \frac{(O-E)^2}{E}$
- Male, A: $(30-20)^2/20=100/20=5$
- Male, B: $(20-30)^2/30=100/30\approx 3.33$
- Female, A: $(10-20)^2/20=100/20=5$
- Female, B: $(40-30)^2/30=100/30\approx 3.33$

$\chi ^2=5+3.33+5+3.33=16.66$


Step 5: Degrees of Freedom

$df=(rows-1)(columns-1)=(2-1)(2-1)=1$


Step 6: Critical Value

At α = 0.05 and df = 1:

$\chi _{critical}^2=3.841$

 Step 7: Decision
- Calculated $\chi ^2=16.66$
- Critical $\chi ^2=3.841$
- Since $16.66>3.841, reject H₀.$


At the 5% significance level, there is strong evidence that product preference is not independent of gender. Gender and product preference are significantly associated.
