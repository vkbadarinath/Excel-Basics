Q1.  Define Covariance and explain how it differs from Correlation in terms of scale and interpretation.

A) - Definition: Covariance measures how two random variables change together.
- Positive covariance → both variables tend to increase or decrease together.
- Negative covariance → one increases while the other decreases.
- Formula:
$\mathrm{Cov}(X,Y)=\frac{\sum (X_i-\bar {X})(Y_i-\bar {Y})}{n-1} $


- Scale: Covariance is not standardized. Its magnitude depends on the units of the variables. For example, if X is measured in kilograms and Y in centimeters, the covariance will be in kg·cm.

 Correlation
- Definition: Correlation is a standardized measure of the strength and direction of the linear relationship between two variables.
- Formula:

  $\mathrm{Corr}(X,Y)=\frac{\mathrm{Cov}(X,Y)}{\sigma _X\cdot \sigma _Y}$

- Scale: Always lies between -1 and +1.
  - +1 → perfect positive linear relationship
  - -1 → perfect negative linear relationship
  - 0 → no linear relationship


**Key Differences:-**

- Scale:

    Covariance: Depends on units of variables

    Correlation: Unit-free, standardized

- Range:

      Covariance: Can be any real number

      Correlation: Always between -1 and +1

- Interpret:

      Covariance:
             Indicates directon and strength of relationship

      Correlation:
             Indicates both direction and strength of relationship

- Usefulness:

      Covariance:
            Hard to compare across datasets

      Correlation:
            Easy to compare across datasets

Q2.  What does a positive, negative, and zero covariance indicate about the relationship between two variables?

A) 🔹 Positive Covariance
- When covariance > 0, the two variables tend to move in the same direction.
- Example: If study hours increase and exam scores also increase, their covariance will be positive.
🔹 Negative Covariance
- When covariance < 0, the two variables tend to move in opposite directions.
- Example: If the amount of time spent watching TV increases while exam scores decrease, their covariance will be negative.
🔹 Zero Covariance
- When covariance ≈ 0, there is no consistent linear relationship between the two variables.
- Example: Shoe size and exam scores would likely have near-zero covariance, since they don’t move together in any meaningful way.

Q3.  Discuss the limitations of covariance as a measure of relationship between two variables. Why is correlation
preferred in many cases?


A)

  **Limitations of Covariance**

Covariance is useful for detecting whether two variables move together, but it has several drawbacks:

1.Scale Dependence
- The magnitude of covariance depends on the units of the variables.
- Example: If you measure height in centimeters vs. meters, the covariance value changes drastically, even though the relationship is the same.

2.Unbounded Values
- Covariance can take any real number (positive or negative), making it hard to interpret the strength of the relationship.
- A covariance of 200 could mean strong or weak association depending on the scale of the data.

3.Comparability Issues
- Because it’s not standardized, you cannot compare covariance values across different datasets or variables.

4.Interpretation Difficulty
- Covariance only tells you the direction (positive or negative) of the relationship, not how strong or consistent.


  **Correlation is Preferred**
Correlation addresses these limitations by standardizing covariance:
- Unit-Free: Correlation removes the effect of measurement units, making results comparable across datasets.
- Bounded Range: Always between -1 and +1, which makes interpretation straightforward.
- Strength + Direction: Correlation clearly indicates both how strong and in which direction the relationship lies.
- Practical Use: Easier to communicate and apply in statistics, finance, machine learning, and research.

Q4. Explain the difference between Pearson’s correlation coefficient and Spearman’s rank correlation
coefficient. When would you prefer to use Spearman’s correlation?

A)

  **Pearson vs. Spearman Correlation**

- Pearson’s Correlation Coefficient
  - Definition: Measures the linear relationship between two continuous variables.
- Assumptions:
  - Variables should be normally distributed (or close to it).
  - Relationship should be linear.
  - Sensitive to outliers (they can distort the coefficient).
- Range: Between -1 and +1.
  - +1 → perfect positive linear relationship
  - -1 → perfect negative linear relationship
  - 0 → no linear relationship
- Spearman’s Rank Correlation Coefficient

- **Definition:** Measures the monotonic relationship between two variables using their ranks rather than raw values.

- **Advantages:**
- Does not require linearity.
- Works well with ordinal data (ranked categories).

Q5.  If the correlation coefficient between two variables X and Y is 0.85, interpret this value in context. Can you
infer causation from this value? Why or why not?

A)

  **Interpreting a Correlation Coefficient of 0.85**
  
- A correlation coefficient (r) of 0.85 indicates a strong positive linear relationship between variables X and Y.
- This means that as X increases, Y tends to increase as well, and the relationship is fairly consistent.
- The closeness to +1 suggests the data points lie near a straight line with a positive slope.

 **Can We Infer Causation?**


- *Correlation only measures association:* It tells us how two variables move together, not whether one causes the other.
- Possible confounding variables: A third factor might be influencing both X and Y.
- Example: Ice cream sales and drowning incidents may be positively correlated, but the underlying cause is hot weather.
- Direction of influence is unclear: Even if there is a causal link, correlation doesn’t tell us whether X causes Y, Y causes X, or both.
- Spurious correlations: Sometimes variables can appear correlated purely by chance, without any meaningful relationship.

A correlation of 0.85 shows a strong positive association, but we cannot conclude causation without further evidence

Q10. Two investment portfolios have the following returns (%) over 5 years. Compute the covariance and
correlation coefficient, and interpret whether the portfolios move together.

A) **Calculate the means**

 $\bar {A}=\frac{8+10+12+9+11}{5}=\frac{50}{5}=10$

 $\bar {B}=\frac{6+9+11+8+10}{5}=\frac{44}{5}=8.8$

 **Compute deviations and products**

 A     B    A-A    B-B    Product

 8     6     -2   -2.8      5.6

 10    9      0    0.2        0

 12   11      2    2.2      4.4

 $\sum (A-\bar {A})(B-\bar {B})=5.6+0+4.4+0.8+1.2=12$

  **Covariance**

$\mathrm{Cov}(A,B)=\frac{\sum (A-\bar {A})(B-\bar {B})}{n-1}$

**Variances**

$\sum (A-\bar {A})^2=(-2)^2+0^2+2^2+(-1)^2+1^2=4+0+4+1+1=10$
